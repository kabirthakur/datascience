```python
import pandas as pd
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from sklearn.preprocessing import LabelEncoder
import networkx as nx
import matplotlib.pyplot as plt
```

### Data import and preprocessing

```python
df=pd.read_csv("bank-data.csv")
```

```python
df.head()
```

```python
str_cols = ['sex', 'region', 'married', 'car', 'save_act', 'current_act', 'mortgage', 'pep']
df = pd.get_dummies(df, columns=str_cols)
```

```python
df
```

```python
def age_separator(age_col):
    age_categories = []
    for age in age_col:
        if age < 25:
            age_categories.append("2")
        elif age < 40:
            age_categories.append("25-40")
        elif age < 55:
            age_categories.append("40-55")
        elif age < 70:
            age_categories.append("55-70")
        else:
            age_categories.append("70+")
    return age_categories
df["age"] = age_separator(df["age"])
```

```python
def income_separator(income_col):
    income_categories = []
    for income in income_col:
        if income < 17000:
            income_categories.append("< 17000")
        elif income < 25000:
            income_categories.append("17000-25000")
        elif income < 35000:
            income_categories.append("25000-35000")
        else:
            income_categories.append("> 35000")
    return income_categories
df["income"] = income_separator(df["income"])
```

```python
df
```

```python
df = pd.concat([df, pd.get_dummies(df["children"], prefix="children")], axis=1)
df = df.drop("children", axis=1)
df = pd.concat([df, pd.get_dummies(df["income"], prefix="income")], axis=1)
df = df.drop("income", axis=1)
df = pd.concat([df, pd.get_dummies(df["age"], prefix="age")], axis=1)
df = df.drop("age", axis=1)
df = df.drop("id",axis=1)
df
```

```python
df = df.astype('bool')
```

### Association Rules Mining

```python
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.5)
rules
```

```python
rules_all = rules[(rules['confidence'] > 0.5) & (rules['lift'] > 1) ].sort_values(by=["lift", "confidence"], ascending=False)
rules_all.head(20)
```

```python
rules_all = rules[(rules['confidence'] > 0.75) & (rules['lift'] > 1.5) ].sort_values(by=["lift", "confidence"], ascending=False)
rules_all
```

```python
rules_with_pep = rules[rules['consequents'] == frozenset({'pep_YES'})]
rules_with_pep = rules_with_pep.sort_values(by=["lift", "confidence"], ascending=False)
rules_with_pep
```

```python
rules_with_pep = rules[rules['consequents'] == frozenset({'pep_YES'})]
rules_with_pep = rules_with_pep[(rules_with_pep['confidence'] > 0.75) & (rules['lift'] > 1.5) ].sort_values(by=["lift", "confidence"], ascending=False)
rules_with_pep
```

### Create a graph

```python
G = nx.DiGraph()

for _, row in rules_with_pep.iterrows():
    
    G.add_edge(', '.join(list(row['antecedents'])), ', '.join(list(row['consequents'])), 
               confidence="{0:.3f}".format(row['confidence']), lift="{0:.3f}".format(row['lift']))

plt.figure(figsize=(12, 12))

pos = nx.spring_layout(G, k=0.5) # k regulates the distance between nodes

edge_labels = dict([((u, v), d['confidence'] + ', ' + d['lift']) for u, v, d in G.edges(data=True)])

nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=1500, edge_color='gray')

nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')

plt.show()
```
