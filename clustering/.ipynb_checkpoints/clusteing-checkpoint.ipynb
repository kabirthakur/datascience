```python
def process_dataframe(df):
    #scaler = StandardScaler()
    scaler= MinMaxScaler()
    for col in df.columns:
        # Check if column needs scaling
        if (df[col].min() < 0) or (df[col].max() > 1):
            # Scale the column
            df[col] = scaler.fit_transform(df[[col]])
            print(col)
    return df
```

```python
df = df_original.drop("author",axis=1)
df = df.drop("filename",axis=1)
df = process_dataframe(df)
```

```python
df
```

### Kmeans

```python
kmeans = KMeans(n_clusters=4)
clusters = kmeans.fit(df)
```

```python
labels = kmeans.labels_
```

### Figuring out the clusters and what authors they have

```python
df_original["Clusters"] = labels
```

```python
df_test = df_original.loc[:,["author","Clusters"]]
```

```python
summary_df = df_test.groupby('Clusters')['author'].value_counts().reset_index(name='Count')

print(summary_df)
```
